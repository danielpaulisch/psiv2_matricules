```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
img = cv.imread('imatge.jpeg')
```

```python
plt.imshow(img)
```

```python
b,g,r = cv.split(img)
```

```python
plt.imshow(b,cmap="gray")
```

```python
plt.imshow(g,cmap="gray")
```

```python
plt.imshow(r,cmap="gray")
```

```python
plt.imshow(cv.merge((r,g,b)))
```

```python
img_gris = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
```

```python
plt.imshow(img_gris,cmap="gray")
```

```python
cv.imwrite("foto_revertida.jpeg",cv.merge((r,g,b)))
```
