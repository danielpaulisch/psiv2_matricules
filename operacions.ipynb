```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
img = cv.imread('matricula_1.jpeg')
img_gris = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
```

```python
img_gris>150
plt.imshow(img_gris>150,cmap="gray")
```

```python
#img = cv.imread('sudoku.png',0)
img_thr = cv.medianBlur(img_gris,5)
th3 = cv.adaptiveThreshold(img_thr,255,cv.ADAPTIVE_THRESH_GAUSSIAN_C,cv.THRESH_BINARY,11,2)
plt.imshow(th3,cmap="gray")
```

```python
blur = cv.GaussianBlur(img_gris,(5,5),0)

# find otsu's threshold value with OpenCV function
ret, otsu = cv.threshold(blur,0,255,cv.THRESH_BINARY+cv.THRESH_OTSU)
```

```python
plt.imshow(otsu,cmap="gray")
```

```python
otsu
```

```python
cnts,_=cv.findContours(cv.Canny(otsu,100,200),cv.RETR_LIST,cv.CHAIN_APPROX_SIMPLE)
contornos=cv.drawContours((otsu//255),cnts,-1,(100,100,100),2)
```

```python
plt.imshow(contornos>50,cmap="gray")
```

```python
kernel=np.ones((3,3),np.uint8)
top_hat=cv.morphologyEx(img_gris,cv.MORPH_BLACKHAT,kernel,iterations=10)
```

```python
plt.imshow(top_hat,cmap="gray")
```
